```python
import pandas as pd
from ai4bharat.transliteration import XlitEngine
```

```python
df = pd.read_csv("data/metadata.txt",sep = "|")
```

```python
ur_rm = pd.read_csv("urdu_to_roman.csv")
ur_rm = ur_rm.set_index('0')['1'].to_dict()
```

```python
e = XlitEngine(src_script_type="indic", beam_width=10, rescore=False)
```

```python
def transliteration(row):
    sentence = row[1]
    words = sentence.split()
    translit = ""
    for word in words:
        try:
            translit += " " + ur_rm[word]
        except:
            translit += " " + e.translit_word(word, lang_code="ur", topk=1)[0]
    if row.name%1000 == 0:
        print(row.name)
    return translit.strip()
```

```python
df['Text'] = df.apply(transliteration, axis=1)
```
