```python
import pandas as pd
```

```python
df = pd.read_csv("data/metadata_roman.txt",sep="|")
```

```python
def change_path(path):
    return '/'.join(path.split('/')[2::2])
```

```python
df['File'] = df['File'].apply(change_path)
```
